```python
# transform MPL to plotly

# plotting as matplotlib method

n = 50
x, y, z, s, ew = np.random.rand(5, n)

c, ec = np.random.rand(2, n, 4)

area_scale, width_scale = 500, 5

fig, ax = plt.subplots()

sc = ax.scatter(x, y, c=c, s=np.square(s)*area_scale,
               edgecolor=ec,
               linewidth=ew*width_scale)

ax.grid()

plotly.offline.plot_mpl(fig)
```
