```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import networkx as nx
import nxviz as nv
import random
```

```python

G=nx.grid_2d_graph(3,10, periodic=True)
nx.draw(G, node_size=30)
plt.show()
_ = nx.draw_networkx(G, with_labels=False, node_size=20, alpha=0.7, width=0.1, node_shape=random.choice(list('so^>v<dph8')), pos=nx.layout.spectral_layout(G))
plt.axis('off')
plt.show()
_ = nx.draw_networkx(G, with_labels=False, node_size=20, alpha=0.7, width=0.1, node_shape=random.choice(list('so^>v<dph8')), pos=nx.layout.spring_layout(G))
plt.axis('off')
plt.show()
_ = nx.draw_networkx(G, with_labels=False, node_size=20, alpha=0.7, width=0.1, node_shape=random.choice(list('so^>v<dph8')), pos=nx.layout.circular_layout(G))
plt.axis('off')
plt.show()
_ = nx.draw_networkx(G, with_labels=False, node_size=20, alpha=0.7, width=0.1, node_shape=random.choice(list('so^>v<dph8')), pos=nx.layout.fruchterman_reingold_layout(G))
plt.axis('off')
plt.show()
_ = nx.draw_networkx(G, with_labels=False, node_size=20, alpha=0.7, width=0.1, node_shape=random.choice(list('so^>v<dph8')), pos=nx.layout.kamada_kawai_layout(G))
plt.axis('off')
plt.show()
_ = nx.draw_networkx(G, with_labels=False, node_size=20, alpha=0.7, width=0.1, node_shape=random.choice(list('so^>v<dph8')), pos=nx.layout.random_layout(G))
plt.axis('off')
plt.show()
_ = nx.draw_networkx(G, with_labels=False, node_size=20, alpha=0.7, width=0.1, node_shape=random.choice(list('so^>v<dph8')), pos=nx.layout.shell_layout(G))
plt.axis('off')
plt.show()
_ = nx.draw_networkx(G, with_labels=False, node_size=20, alpha=0.7, width=0.1, node_shape=random.choice(list('so^>v<dph8')), pos=nx.layout.spring_layout(G))
plt.axis('off')
plt.show();
```
